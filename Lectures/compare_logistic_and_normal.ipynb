```python
from scipy.stats import logistic
from scipy.stats import norm
import matplotlib.pyplot as plt

samples_log = logistic().rvs(size=1000)
samples_norm = norm().rvs(size=1000)

plt.hist(samples_log, color='blue', label='logistic')
plt.hist(samples_norm, color='yellow', label='standard normal')
plt.legend()
plt.show()
```
